scatter plot

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
movie=['batman','avengers','ironman','mask']
percentage=[23,49,47,60]
```

```python
plt.scatter(movie,percentage)
```

line plot

```python
plt.plot(movie,percentage,linestyle='dashdot')
```

```python
plt.plot(movie,percentage,linestyle='dotted')
```

```python
plt.plot(movie,percentage,linestyle='dashed')
```

bar chat

```python
plt.bar(movie,percentage)
```

pie chart

```python
plt.pie(percentage,labels=movie)
```
